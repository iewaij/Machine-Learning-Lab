```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from __future__ import division
import seaborn as sns
%pylab inline
```

```python
x1, x2, y = np.loadtxt("data.csv", delimiter=',', usecols=(0, 1, 2), unpack=True)
X = np.vstack((x1, x2))
X = X.astype(np.float32)
y = y.astype(np.float32)
print X.shape

for n in range(200):
    if y[n] == 0.0:
        plt.plot(X[0][n], X[1][n], '.', c='#1E88A8')
    elif y[n] == 1.0:
        plt.plot(X[0][n], X[1][n], '.', c='#D0104C')
```

```python
#Sample, size = k
#从数据中抽样，样本大小为 k
def sample(k):
    index = np.random.randint(200-k)
    X_rd = X[:, index:index+k]
    y_rd = y[index:index+k]
    return X_rd, y_rd
```

```python
def generate_weight_and_bias(input_dim):
    # w, 1 x input_dim
    # 行向量，也可视为一维矩阵
    w = tf.Variable(tf.random_uniform([1, input_dim], -1, 1))
    # b, 1 * 1
    bias = tf.Variable(tf.zeros([1, 1]))
    return (w, bias)

def accuracy(preds, labels):
    return ((preds > 0.5) == labels).sum() / float(len(labels))
```

```python
tf.reset_default_graph()
W, b = generate_weight_and_bias(2)
X_ph = tf.placeholder('float')
y_ph = tf.placeholder('float')
nn_raw_output = tf.add(tf.matmul(W, X_ph), b)
final_output = tf.sigmoid(nn_raw_output)
```

```python
cost = tf.reduce_mean(
    tf.nn.sigmoid_cross_entropy_with_logits(logits=nn_raw_output, labels=y_ph)
)
```

```python
x1_axis = np.linspace(-2, 2, 100)
x1_axis, x2_axis = np.meshgrid(x1_axis, x1_axis)
X_axis = np.c_[x1_axis.ravel(), x2_axis.ravel()].T.astype(np.float32)
```

```python
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cost)
with tf.Session() as sess:
    feed_dict = {X_ph:X,y_ph:y}
    sess.run(tf.global_variables_initializer())
    for i in range(500):
        sess.run(train_step, feed_dict = feed_dict)
        if i % 50 == 0:
            Z_axis = sess.run(final_output, {X_ph:X_axis,y_ph:y})
            f, axarr = plt.subplots(sharex='col', sharey='row')
            axarr.contourf(x1_axis, x2_axis, Z_axis.reshape(x1_axis.shape), alpha = 0.4)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure('3D scatter plot')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1_axis, x2_axis, Z_axis.reshape(x1_axis.shape))
```
